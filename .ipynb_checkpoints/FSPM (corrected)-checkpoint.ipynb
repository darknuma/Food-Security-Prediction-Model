# The Skew and the Kurtosis

<b>SKEW</b>

	A negative skew indicates that the tail is on the left side of the distribution, which extends towards more negative values.

	A positive skew indicates that the tail is on the right side of the distribution, which extends towards more positive values.

	A value of zero indicates that there is no skewness in the distribution at all, meaning the distribution is perfectly symmetrical.


```python
df.skew(axis = 0, skipna = True)
```

<b>Kurtosis</b> 
<br>
A positive value tells you that you have heavy-tails (i.e. a lot of data in your tails).

A negative value means that you have light-tails (i.e. little data in your tails).

```python
df.kurtosis(axis=0)
```

```python
boxcox_trans = pd.DataFrame()
boxcox_trans['POU'] = stats.boxcox(df['Prevalence of undernourishment (percent) (3-year average) %'])[0]
boxcox_trans['CIDR'] = stats.boxcox(df['Cereal import dependency ratio (percent) (3-year average)'])[1]
boxcox_trans['VFI'] = stats.boxcox(df['Value of food imports in total merchandise exports (percent) (3-year average) %'])[0]

boxcox_trans['LEI'] = stats.boxcox(df['Percent of arable land equipped for irrigation (percent) (3-year average) %'])[0]
boxcox_trans['ADES'] = stats.boxcox(df['Average dietary energy supply adequacy (percent) (3-year average) %'])[1]
boxcox_trans['SDES'] = stats.boxcox(df['Share of dietary energy supply derived from cereals, roots and tubers (kcal/cap/day) (3-year average)'])[0]

boxcox_trans['SMDS'] = stats.boxcox(df['Percentage of population using safely managed drinking water services (Percent)'])[0]
boxcox_trans['LDS'] = stats.boxcox(df['Percentage of population using at least basic drinking water services (percent)'])[1]
boxcox_trans['POA'] = stats.boxcox(df['Prevalence of obesity in the adult population (18 years and older)'])[0]

boxcox_trans['GDP'] = stats.boxcox(df['Gross domestic product per capita, PPP, dissemination (constant 2011 international $)'])[0]
boxcox_trans['FSV'] = stats.boxcox(df['Per capita food supply variability (kcal/cap/day)'])[0]

# Political Instability is Negative and BoxCox Transformation won't work for it
# boxcox_trans['PSAVT'] = stats.boxcox(df['Political stability and absence of violence/terrorism (index)'])[0]

boxcox_trans
```

```python
df_tab.columns = ['POU', 'CIDR',
                        'VFI', 'LEI',
                        'ADES', 'SDES',
                        'SMDS', 'LDS',
                        'POA', 'GDP',
                        'FSV']
df_tab
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df_tab.iloc[:,:-1]
calc_vif(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sc_normalized,y,test_size=0.15)
```

```python
from sklearn import linear_model
from sklearn import svm
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

classifiers = [
    svm.SVR(),
    linear_model.SGDRegressor(),
    linear_model.Ridge(),
    linear_model.BayesianRidge(),
    linear_model.LassoLars(),
    linear_model.ARDRegression(),
    linear_model.PassiveAggressiveRegressor(),
    linear_model.TheilSenRegressor(),
    linear_model.LinearRegression()]
for item in classifiers:
    print(item)
    clf = item
    clf.fit(X_train, y_train)
    # print(clf.predict(X_test),'\n')
    y_pred = clf.predict(X_test)
    y_score = clf.score(X_test, y_test)
    print(y_pred, '\n')
    print("The mean absolute error {}".format(mean_absolute_error(y_test, y_pred),'\n'))
    print("The mean square error {}".format(mean_squared_error(y_test, y_pred),'\n'))
    print("The r2 score {}".format(r2_score(y_test, y_pred),'\n'))
   
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
results = [['Model', 'MAE']]


rndm_frst = RandomForestRegressor(n_estimators=300, random_state=0)

rndm_frst.fit(X_train, y_train)

y_pred = rndm_frst.predict(X_test)

rms = sqrt(mean_squared_error(y_pred, y_test))

print("MAE: " + str(mean_absolute_error(y_pred, y_test)))
print("MSE: " + str(mean_squared_error(y_pred, y_test)))
print("r2_score: " + str(r2_score(y_pred, y_test)))
print("RMSE: " + str(rms))

results.append(['Random Forest Regressor', str(mean_absolute_error(y_pred, y_test))])
results.append(['Random Forest Regressor', str(mean_squared_error(y_pred, y_test))])
results.append(['Random Forest Regressor', str(r2_score(y_pred, y_test))])
results.append(['Random Forest Regressor', str(rms)])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

X = food_availabilty.iloc[:,:]
calc_vif(X)
```
